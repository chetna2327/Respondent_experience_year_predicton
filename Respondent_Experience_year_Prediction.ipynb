```python
# to create a model that can predict the number of years' experience a
# respondent has had programming.


df_new =pd.DataFrame(df[["MainBranch","YearsCodePro","MiscTechDesireNextYear","NEWCollabToolsWorkedWith","NEWLearn","DatabaseWorkedWith","DatabaseWorkedWith","WebframeWorkedWith",
]])
df_new
```

```python
df_new.isna().sum()
```

```python
df1=df_new.dropna()
df1
```

```python
"""spliting the values as columns from "MainBranch" column"""

list_of_branches=list(set([y.lower().strip() for x in df1["MainBranch"] for y in x.split(",")]))
features_main_branch=pd.DataFrame([[0]*len(list_of_branches)]*df1["MainBranch"].shape[0],columns=list_of_branches)
for i,x in enumerate(df1["MainBranch"]):
    for y in x.split(","):
        yy=y.lower().strip()
        if yy in list_of_branches:
            features_main_branch.loc[i,yy]=1
```

```python
y_df = df1[["MainBranch"]]
```

```python
df2=pd.concat([features_main_branch,df1], axis=1)
```

```python
df2 = df2.drop('MainBranch',1)
df2 = df2.drop('MiscTechDesireNextYear',1)
df2 = df2.drop('NEWCollabToolsWorkedWith',1)
df2 = df2.drop('DatabaseWorkedWith',1)
df2 = df2.drop('WebframeWorkedWith',1)


```

```python
df2
```

```python
df3= df2
```

```python
df3["NEWLearn"]=df3["NEWLearn"].factorize()[0]
df3["YearsCodePro"]=df3["YearsCodePro"].factorize()[0]

```

```python
df3
```

```python
df3["YearsCodePro"].describe()
```

```python
df4 = df3.dropna()
```

```python
""" spliting data into test and training data"""

from sklearn.model_selection import train_test_split  
X_train,X_test,Y_train, Y_test = train_test_split(df4, df4['YearsCodePro'], test_size=0.3, random_state=10)
print("Train: ",X_train.shape,Y_train.shape,"Test: ",(X_test.shape,Y_test.shape))
```

```python
""" importing Models from sklearn"""

from sklearn.metrics import mean_absolute_percentage_error
def eval_metrics(rfr,X_test,Y_test):
    print("MAPE: ",100-mean_absolute_percentage_error(Y_test, rfr.predict(X_test)))

    from sklearn.metrics import median_absolute_error
    print("MEDIAN ABSOLUTE: ",100-median_absolute_error(Y_test, rfr.predict(X_test)))


    from sklearn.metrics import r2_score
    print("R2: ",r2_score(Y_test, rfr.predict(X_test))*100)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
```

```python
lr=LinearRegression()
lr.fit(X_train, Y_train)
eval_metrics(lr,X_test,Y_test)
```

```python
ridg=Ridge()
ridg.fit(X_train, Y_train)
eval_metrics(ridg,X_test,Y_test)
```

```python
rfr = RandomForestRegressor(n_estimators=15, random_state=100)
rfr.fit(X_train, Y_train) 
eval_metrics(rfr,X_test,Y_test)
```
